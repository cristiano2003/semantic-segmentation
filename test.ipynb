```python
from src.neopolyp.dataset.coco_utils import *

dataset = infer_build("val")

dataset[0]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have the dataset and sample as defined in the previous example
image, mask = dataset[0][0], dataset[0][1]
image = torch.permute(image, (1, 2, 0))
# Display the RGB image
plt.subplot(1, 2, 1)
plt.imshow(np.array(image))
plt.title('RGB Image')
plt.axis('off')

# Display the mask
plt.subplot(1, 2, 2)
plt.imshow(np.array(mask), cmap='jet')  # You can choose a different colormap
plt.title('Mask')
plt.axis('off')

plt.show()
```
